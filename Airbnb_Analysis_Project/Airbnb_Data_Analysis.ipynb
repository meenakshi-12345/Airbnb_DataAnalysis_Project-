```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# **Checking for missing values in dataset**

```python
print(df.isnull().sum())
```

```python
df.info()
```

# **Handling missing values**

```python
df['last review'] = pd.to_datetime(df['last review'], errors = 'coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month' : 0, 'last review' : df['last review'].min()}, inplace = True)
```

```python
df.dropna(subset =  ['NAME', 'host name'], inplace = True)
```

```python
print(df.isnull(). sum())
```

```python
df.drop(columns = ['license', 'house_rules'], errors = 'ignore')

```

```python
df['price'] = df['price'].replace('[\$,]', '', regex = True).astype(float)

df['service fee'] = df['service fee'].replace('[\$,]', '', regex = True).astype(float)
```

```python
df.head()
```

# **Removing Duplicates**

```python
df.drop_duplicates(inplace = True)
```

```python
df.info()
```

```python
df.describe()
```

# **Visualizating Dataset and finding useful insights**

*What is the Distribution of listing prices?*

```python
plt.figure(figsize= (10,6))

sns.histplot(df['price'], bins = 50, kde = True, color = 'red')

plt.title("Distribution of Listing Price")

plt.xlabel("Price $")

plt.ylabel("Frequency")

plt.show()
```

The histogram shows a fairly even distribution of listing prices across different price ranges, indication no particular concentration of listings in any specific price range. The KDE line helps visualise this even spread more clearly, confirming that the dataset contains listings with a wide variety of prices

*How are different room types distributed?*

```python
df['room type']
```

```python
plt.figure(figsize = (8,5))

sns.countplot(x = 'room type', data = df, color = 'hotpink')

plt.title("Room Type Distribution")

plt.xlabel("Room Type")

plt.ylabel("Count")

plt.show()
```

The most booked room type is entire home/apt followed by private room. The least booked room type is hotel room

*How are listings distributes across different neighbourhoods? *

```python
plt.figure(figsize = (8,6))

sns.countplot( y = "neighbourhood group", data = df, color = 'lightgreen', order= df['neighbourhood group']. value_counts().index)

plt.title("Number Of Listings By Neighborhood Group")

plt.xlabel("Count")

plt.ylabel("Neighbourhood Group")

plt.show()
```

Manhattan leads in listing by nighbourhood group followed by brooklyn and queens

*What is the relationship between price and room type?*

```python
plt.figure(figsize=(8,6))

sns.boxplot(x='room type', y='price', hue = 'room type', data =df, palette = 'Set1')

plt.title("Price Vs Room Type")

plt.xlabel("Room Type")

plt.ylabel("Price ($)")

plt.legend(title = "Room Type")

plt.show()
```

The relationship is shown clearly between Room Type and their prices

*How has the number of reviews over with time?*

```python
df['last reviews'] = pd.to_datetime(df['last review'])

reviews_over_time = df.groupby(df['last review']. dt.to_period('M')).size()

plt.figure(figsize=(10,6))

reviews_over_time.plot(kind = "line", color ="blue")

plt.title("Number Of Reviews Over Time")

plt.xlabel("Date")

plt.ylabel("Number Of Reviews")

plt.show()
```

Review given heavly in the year between 2019 to 2024
